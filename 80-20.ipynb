```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset
file_path = '/kaggle/input/final25k-8020/final25kV2.csv'
df = pd.read_csv(file_path)
df.head(5)
```

```python
empty_indexes = df[df['Category'].isnull()].index

# Print the indexes
print(empty_indexes)
```

```python
print(df['Category'].value_counts())
```

```python
# Shuffle the dataset
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Separate samples with label 1 and label 0
error_df = df[df['Error'] == 1]
correct_df = df[df['Error'] == 0]

# Split each category into train and test maintaining 80-20 ratio
train_error, test_error = train_test_split(error_df, test_size=0.2, stratify=error_df['Category'], random_state=42)
train_correct, test_correct = train_test_split(correct_df, test_size=0.2, stratify=correct_df['Category'], random_state=42)

# Concatenate the train and test sets
train_set = pd.concat([train_error, train_correct], ignore_index=True)
test_set = pd.concat([test_error, test_correct], ignore_index=True)

# Save the train and test sets to CSV files
train_set.to_csv('train8020.csv', index=False)
test_set.to_csv('test8020.csv', index=False)
 
```
